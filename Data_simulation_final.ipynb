```python
import numpy as np
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm, bernoulli
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

# Define Asia Network structure
def create_asia_network():
    G = nx.DiGraph()
    nodes = ['Visit_to_Asia', 'Tuberculosis', 'Smoking', 'Cancer',
             'TB_or_Cancer', 'X_ray', 'Dyspnea', 'Bronchitis']
    G.add_nodes_from(nodes)
    edges = [
        ('Visit_to_Asia', 'Tuberculosis'),
        ('Smoking', 'Cancer'),
        ('Smoking', 'Bronchitis'),
        ('Tuberculosis', 'TB_or_Cancer'),
        ('Cancer', 'TB_or_Cancer'),
        ('TB_or_Cancer', 'X_ray'),
        ('Bronchitis', 'Dyspnea'),
        ('TB_or_Cancer', 'Dyspnea')
    ]
    G.add_edges_from(edges)
    return G

def generate_discrete_data(n_samples):
    # Random probabilities for each node
    probs = {
        'Visit_to_Asia': np.random.uniform(0.01, 0.1),
        'Smoking': np.random.uniform(0.2, 0.5),
    }

    data = pd.DataFrame()

    # Generate root nodes
    data['Visit_to_Asia'] = bernoulli.rvs(probs['Visit_to_Asia'], size=n_samples)
    data['Smoking'] = bernoulli.rvs(probs['Smoking'], size=n_samples)

    # Generate dependent nodes with conditional probabilities
    data['Tuberculosis'] = np.where(
        data['Visit_to_Asia'] == 1,
        bernoulli.rvs(np.random.uniform(0.05, 0.2), size=n_samples),
        bernoulli.rvs(np.random.uniform(0.01, 0.05), size=n_samples)
    )

    data['Cancer'] = np.where(
        data['Smoking'] == 1,
        bernoulli.rvs(np.random.uniform(0.1, 0.3), size=n_samples),
        bernoulli.rvs(np.random.uniform(0.01, 0.1), size=n_samples)
    )

    data['TB_or_Cancer'] = np.where(
        (data['Tuberculosis'] == 1) | (data['Cancer'] == 1),
        1, 0
    )

    data['X_ray'] = np.where(
        data['TB_or_Cancer'] == 1,
        bernoulli.rvs(np.random.uniform(0.7, 0.9), size=n_samples),
        bernoulli.rvs(np.random.uniform(0.05, 0.2), size=n_samples)
    )

    data['Bronchitis'] = np.where(
        data['Smoking'] == 1,
        bernoulli.rvs(np.random.uniform(0.4, 0.7), size=n_samples),
        bernoulli.rvs(np.random.uniform(0.1, 0.3), size=n_samples)
    )

    data['Dyspnea'] = np.where(
        (data['TB_or_Cancer'] == 1) | (data['Bronchitis'] == 1),
        bernoulli.rvs(np.random.uniform(0.7, 0.9), size=n_samples),
        bernoulli.rvs(np.random.uniform(0.1, 0.3), size=n_samples)
    )
    base_filename="discrete"
    param_value=n_samples
    filename = f"{base_filename}_{param_value}.csv"


    data.to_csv(filename, index=True)

    return data

def generate_continuous_data(n_samples):
    data = pd.DataFrame()

    # Generate root nodes with normal distributions
    data['Visit_to_Asia'] = norm.rvs(loc=0, scale=1, size=n_samples)
    data['Smoking'] = norm.rvs(loc=1, scale=2, size=n_samples)

    # Generate dependent nodes with conditional normal distributions
    noise = np.random.normal(0, 0.1, n_samples)
    data['Tuberculosis'] = 0.7 * data['Visit_to_Asia'] + noise

    noise = np.random.normal(0, 0.1, n_samples)
    data['Cancer'] = 0.8 * data['Smoking'] + noise

    data['TB_or_Cancer'] = data['Tuberculosis'] + data['Cancer'] + noise

    noise = np.random.normal(0, 0.1, n_samples)
    data['X_ray'] = 0.9 * data['TB_or_Cancer'] + noise

    noise = np.random.normal(0, 0.1, n_samples)
    data['Bronchitis'] = 0.6 * data['Smoking'] + noise

    noise = np.random.normal(0, 0.1, n_samples)
    data['Dyspnea'] = 0.5 * data['TB_or_Cancer'] + 0.5 * data['Bronchitis'] + noise
    base_filename="continuous"
    param_value=n_samples
    filename = f"{base_filename}_{param_value}.csv"


    data.to_csv(filename, index=True)

    return data

def generate_mixed_data(n_samples):
    """
    Generate mixed dataset with both discrete and continuous variables.
    Smoking and Dyspnea will be continuous, rest will be discrete.
    """
    # First generate all discrete data
    data = pd.DataFrame()

    # Generate root nodes
    data['Visit_to_Asia'] = bernoulli.rvs(np.random.uniform(0.01, 0.1), size=n_samples)
    # Initially discrete Smoking
    data['Smoking_discrete'] = bernoulli.rvs(np.random.uniform(0.2, 0.5), size=n_samples)

    # Generate other discrete nodes
    data['Tuberculosis'] = np.where(
        data['Visit_to_Asia'] == 1,
        bernoulli.rvs(np.random.uniform(0.05, 0.2), size=n_samples),
        bernoulli.rvs(np.random.uniform(0.01, 0.05), size=n_samples)
    )

    data['Cancer'] = np.where(
        data['Smoking_discrete'] == 1,
        bernoulli.rvs(np.random.uniform(0.1, 0.3), size=n_samples),
        bernoulli.rvs(np.random.uniform(0.01, 0.1), size=n_samples)
    )

    data['TB_or_Cancer'] = np.where(
        (data['Tuberculosis'] == 1) | (data['Cancer'] == 1),
        1, 0
    )

    data['X_ray'] = np.where(
        data['TB_or_Cancer'] == 1,
        bernoulli.rvs(np.random.uniform(0.7, 0.9), size=n_samples),
        bernoulli.rvs(np.random.uniform(0.05, 0.2), size=n_samples)
    )

    data['Bronchitis'] = np.where(
        data['Smoking_discrete'] == 1,
        bernoulli.rvs(np.random.uniform(0.4, 0.7), size=n_samples),
        bernoulli.rvs(np.random.uniform(0.1, 0.3), size=n_samples)
    )

    # Initially discrete Dyspnea
    data['Dyspnea_discrete'] = np.where(
        (data['TB_or_Cancer'] == 1) | (data['Bronchitis'] == 1),
        bernoulli.rvs(np.random.uniform(0.7, 0.9), size=n_samples),
        bernoulli.rvs(np.random.uniform(0.1, 0.3), size=n_samples)
    )

    # Convert Smoking to continuous
    # Use normal distribution centered around the discrete value
    data['Smoking'] = np.where(
        data['Smoking_discrete'] == 1,
        norm.rvs(loc=1, scale=0.2, size=n_samples),  # Centered around 1 for smokers
        norm.rvs(loc=0, scale=0.2, size=n_samples)   # Centered around 0 for non-smokers
    )

    # Convert Dyspnea to continuous using logistic regression
    # First, create features for logistic regression
    X = data[['TB_or_Cancer', 'Bronchitis']]
    y = data['Dyspnea_discrete']

    # Fit logistic regression
    lr = LogisticRegression(random_state=42)
    lr.fit(X, y)

    # Get probabilities and add noise
    probs = lr.predict_proba(X)[:, 1]
    data['Dyspnea'] = probs + norm.rvs(loc=0, scale=0.1, size=n_samples)

    # Cleanup: remove temporary columns
    data = data.drop(['Smoking_discrete', 'Dyspnea_discrete'], axis=1)

    # Ensure all values are within reasonable ranges
    data['Smoking'] = np.clip(data['Smoking'], 0, 2)
    data['Dyspnea'] = np.clip(data['Dyspnea'], 0, 1)
    base_filename="mixed"
    param_value=n_samples
    filename = f"{base_filename}_{param_value}.csv"


    data.to_csv(filename, index=True)

    return data

# Example usage and verification
def verify_mixed_data(n_samples=1000):
    """
    Generate and verify the mixed dataset
    """
    data = generate_mixed_data(n_samples)

    print("\nDataset Info:")
    print(data.info())

    print("\nSummary Statistics:")
    print(data.describe())

    print("\nVariable Types:")
    for column in data.columns:
        print(f"{column}: {data[column].dtype}")
        print(f"Unique values: {len(data[column].unique())}")
        print(f"Range: [{data[column].min():.2f}, {data[column].max():.2f}]")
        print("---")

    return data

def visualize_network(G, data, title):
    plt.figure(figsize=(12, 8))

    # Use hierarchical layout instead of spring layout
    pos = nx.kamada_kawai_layout(G)

    # Determine continuous and discrete variables
    continuous_vars = []
    discrete_vars = []
    for col in data.columns:
        if len(data[col].unique()) > 2:
            continuous_vars.append(col)
        else:
            discrete_vars.append(col)

    # Draw nodes with different colors and shapes
    nx.draw_networkx_nodes(G, pos,
                          nodelist=discrete_vars,
                          node_color='lightblue',
                          node_shape='o',
                          node_size=3000,
                          alpha=0.7)

    nx.draw_networkx_nodes(G, pos,
                          nodelist=continuous_vars,
                          node_color='lightgreen',
                          node_shape='s',
                          node_size=3000,
                          alpha=0.7)

    # Draw edges with arrows
    nx.draw_networkx_edges(G, pos,
                          arrows=True,
                          arrowsize=20,
                          edge_color='gray',
                          width=2,
                          arrowstyle='->',
                          connectionstyle='arc3,rad=0.2')

    # Add labels with white background for better visibility
    labels = {}
    for node in G.nodes():
        labels[node] = node.replace('_', '\n')

    # Create label backgrounds
    bbox_props = dict(boxstyle="round,pad=0.3", fc="white", ec="gray", alpha=0.7)
    nx.draw_networkx_labels(G, pos, labels, font_size=10, font_weight='bold', bbox=bbox_props)

    # Add legend
    discrete_patch = plt.Rectangle((0, 0), 1, 1, fc="lightblue", alpha=0.7)
    continuous_patch = plt.Rectangle((0, 0), 1, 1, fc="lightgreen", alpha=0.7)
    plt.legend([discrete_patch, continuous_patch],
              ['Discrete Variables', 'Continuous Variables'],
              loc='upper left',
              bbox_to_anchor=(1, 1))

    plt.title(title, pad=20, size=14, fontweight='bold')
    plt.axis('off')
    # Add more space around the graph
    plt.tight_layout()
    plt.margins(0.2)
    plt.show()

def analyze_dataset(data, dataset_type):
    print(f"\nAnalysis for {dataset_type} dataset:")
    print("\nSummary Statistics:")
    print(data.describe())

    print("\nCorrelation Analysis:")
    print(data.corr().round(2))

    # Distribution plots for each variable
    plt.figure(figsize=(15, 10))
    for i, col in enumerate(data.columns, 1):
        plt.subplot(3, 3, i)
        if len(data[col].unique()) <= 2:  # Discrete variable
            sns.countplot(data=data, x=col)
        else:  # Continuous variable
            sns.histplot(data=data, x=col, kde=True)
        plt.title(f'Distribution of {col}')
    plt.tight_layout()
    plt.show()

# Main execution block
def main():
    # Generate datasets for different sample sizes
    sample_sizes = [10, 100, 1000, 10000]
    G = create_asia_network()

    for size in sample_sizes:
        print(f"\n=== Sample Size: {size} ===")

        # Generate and visualize discrete dataset
        discrete_data = generate_discrete_data(size)
        visualize_network(G, discrete_data, f"Discrete Dataset (n={size})")
        analyze_dataset(discrete_data, "Discrete")

        # Generate and visualize continuous dataset
        continuous_data = generate_continuous_data(size)
        visualize_network(G, continuous_data, f"Continuous Dataset (n={size})")
        analyze_dataset(continuous_data, "Continuous")

        # Generate and visualize mixed dataset
        mixed_data = generate_mixed_data(size)
        visualize_network(G, mixed_data, f"Mixed Dataset (n={size})")
        analyze_dataset(mixed_data, "Mixed")

if __name__ == "__main__":
    main()
```
